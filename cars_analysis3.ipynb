```python
#To find the no of cars based on their launch years.


# Group the data by car launch year and count the occurrences
year_counts = df['year'].value_counts().sort_index()                 #this will take distinct years in ascending order and count it

# Extract the years and counts
years = year_counts.index        #this will take all the distinct years
counts = year_counts.values      #this will take the values from it i.e number of vehicles

# Create a line chart
plt.figure(figsize=(10, 5))                             #to set the figure size
plt.plot(years, counts, marker='o', linestyle='-')      #this will create a line chart based on mentioned axes

# Set labels and title
plt.xlabel("Car Launch Year")                #to set display text on x axis
plt.ylabel("Count of Cars")                  #to set display text on y axis
plt.title("Count of Cars by Launch Year")    #to set title of the line chart

# Show the plot
plt.tight_layout()      #this will adjust the space between plot
plt.show()              #used to show the above created line chart


'''Analysis
We have less number of cars of old model that is 1980 to 2004.
But we have more number of cars from 2005 to 2019.'''
```

```python
#Find the no of cars based on their selling price.


# Extract the 'selling_price' column
selling_prices = df['selling_price']

# Create the histogram
plt.figure(figsize=(15,6))                                             #to set the size of the plot
plt.hist(selling_prices, bins=50, color='blue', rwidth=0.85)           #to make a histogram based on mentioned column, rwidth is space between bars of histogram
plt.xlabel('Selling Price')                                            #to display text on x axis 
plt.ylabel('Number of Cars')                                           #to display text on y axis
plt.title('Selling Prices for the Available Cars')                 #to display text on title

plt.tight_layout()        #to adjust the space between the plots
plt.show()                #to display the created histogram


'''Analysis 
By observing the plot we got to know that we have large number of cars of low prices and small number of cars of high prices'''
```

```python
#To compare two quantiative values that is Selling price and Count of cars. 


# Group the data by 'selling_price' and count the number of cars at each price
price_counts = df['km_driven'].value_counts().reset_index()              #to count different all the values of columns
price_counts.columns = ['Selling Price', 'Count']

plt.figure(figsize=(10,6))         #to set the size of the plot
plt.scatter(price_counts['Selling Price'], price_counts['Count'], alpha=0.5)     #to make the scatter plot and alpha is for opacity adjustment
plt.xlabel('Selling Price')        #to set text on x axis
plt.ylabel('Count of Cars')        #to set text on y axis
plt.title('Scatter Plot of Selling Price vs. Count of Cars')      #to set text on title

plt.tight_layout()         #to adjust the space between the plots
plt.show()                 #to display the created histogram


'''Analysis
Here we got to know that we have maximum cars with the medium and small price range so we can say that we have lot of cars for middle class people.
But we have more parameters to analyze such as car overall performance, lets take a look into it. '''
```

```python
#To check the distribution of three quantitative columnns that is engine, maximum power and mileage.


columns_to_plot = ['engine', 'max_power', 'mileage']      #list of columns to make the boxplot

# Create a boxplot for each selected column
plt.figure(figsize=(12, 6))            #to set the size of the plot

# Loop through the columns and create side-by-side boxplots
for i, column in enumerate(columns_to_plot, start=1):           #using loop with enumerate traversing the list of columns 
    plt.subplot(1, len(columns_to_plot), i)             #to make multiple plots
    plt.boxplot(df[column], boxprops={'color': 'blue'}, whiskerprops={'color': 'red'}, medianprops={'color': 'green'}, flierprops={'marker': 'o', 'markerfacecolor': 'pink', 'markeredgecolor': 'black'})               #to make boxplot with all the appearance properties
    plt.title(column)         #to set title of all the columns based on their names
    plt.ylabel('Distribution')     #to set text on title

plt.tight_layout()      #to adjust the space between the plots
plt.show()              #to display the created histogram


'''Analysis
Here we have engine ranges from 500 CC to 3600 CC.
Maximum power from 60-70 bhp to 450+ bhp.
Mileage from 10 to 42 kmpl.
Here no outliers are considered, all the data are important to consider for analysis.'''
```

Bivariate Analysis

```python
#To find the price of the cars based on their body types and compare them.


plt.figure(figsize=(20,13))     #to set the size of the plot
plt.subplot(2,2,1)              #to make 2,2 plot so to insert multiple plots in one
# Group the data by 'car_body_type' and calculate the average selling price for each type
avg_price_by_body_type = df.groupby('body_type')['selling_price'].mean().reset_index()

# Create the bar graph
plt.bar(avg_price_by_body_type['body_type'], avg_price_by_body_type['selling_price'])       #to create a bar graph
plt.xlabel('Car Body Type')                              #to set text on x axis                          
plt.ylabel('Average Selling Price')                      #to set text on y axis
plt.title('Average Selling Price by Car Body Type')      #to set text on title


#another graph, which shows highest selling price.
plt.subplot(2,2,2)
# Group the data by 'car_body_type' and calculate the average selling price for each type
max_price_by_body_type = df.groupby('body_type')['selling_price'].max().reset_index()

# Create the bar graph
plt.plot(max_price_by_body_type['body_type'], max_price_by_body_type['selling_price'], marker='o')
plt.xlabel('Car Body Type')
plt.ylabel('Highest Selling Price')
plt.title('Highest Selling Price by Car Body Type')


#another graph, which shows lowest selling price.
plt.subplot(2,2,3)
# Group the data by 'car_body_type' and calculate the average selling price for each type
min_price_by_body_type = df.groupby('body_type')['selling_price'].min().reset_index()

# Create the bar graph
plt.plot(min_price_by_body_type['body_type'], min_price_by_body_type['selling_price'], marker='o')
plt.xlabel('Car Body Type')
plt.ylabel('Lowest Selling Price')
plt.title('Lowest Selling Price by Car Body Type')


#plt.tight_layout()      #to adjust the space between the plots
plt.show()              #to display the created graphs


'''Analysis
Average selling price of Coupe cars are highest.
SUVs has a car with highest selling price i.e, Rs - 10000000
Convertible, Coupe and Sedan has cars with lowest selling price i.e, Rs - 30000'''
```

```python
#To find out the average selling price of cars based on their launch year.


plt.figure(figsize=(10,6))      #to set the size of the plot

# Group the data by 'car_body_type' and calculate the average selling price for each type
avg_price_by_launch_year = df.groupby('year')['selling_price'].mean().reset_index()

# Create the bar graph
plt.bar(avg_price_by_launch_year['year'], avg_price_by_launch_year['selling_price'])      #to make the bar graph
plt.xlabel('Car Launch Year')                              #to set text on x axis
plt.ylabel('Average Selling Price')                        #to set text on y axis
plt.title('Average Selling Price by Car Launch Year')      #to set text on title

plt.tight_layout()      #to adjust the space between the plots
plt.show()              #to display the created bar graph


'''Analysis
Here we got to know that the car with launch year before 2010 have low selling price.
After that the average price is gradually rising.
Cars of launch year around 2019 have the highest selling price that is above Rs 1750000.'''
```

```python
#To compare all the car types on the basis of Engine, Power and Mileage.


plt.figure(figsize=(30,12))      #to set the size of the plot
plt.subplot(3,3,1)               #to create 3,3 matrix so that we can insert multiple plots

# Calculate the average, maximum, and minimum engine for each car body type
agg_df1 = df.groupby("body_type")["engine"].agg(['mean', 'max', 'min']).reset_index()

# Create the side-by-side bar graph
bar_width = 0.2                #to set width of the bar
index = range(len(agg_df1))     #according to this placing will be done of bars

plt.bar(index, agg_df1['mean'], bar_width, label='Average Engine', color='b', align='center')                             #to make first bar of average engine
plt.bar([i + bar_width for i in index], agg_df1['max'], bar_width, label='Max Engine', color='g', align='center')         #to make first bar of max engine
plt.bar([i + 2 * bar_width for i in index], agg_df1['min'], bar_width, label='Min Engine', color='r', align='center')     #to make first bar of min engine

# Customize the graph
plt.xlabel('Car Body Type')      #to set text on x axis
plt.ylabel('Engine(in CC)')      #to set text on y axis
plt.title('Engine Statistics by Car Body Type')         #to set text on title
plt.xticks([i + bar_width for i in index], agg_df1['body_type'])   #to locate car body type on x axis that is in the middle
plt.legend()         #to show detail of bars


#to make the second graph that is Comparision of Power
plt.subplot(3,3,4)
# Calculate the average, maximum, and minimum power for each car body type
agg_df2 = df.groupby("body_type")["max_power"].agg(['mean', 'max', 'min']).reset_index()

# Create the side-by-side bar graph
bar_width = 0.2
index = range(len(agg_df2))

plt.bar(index, agg_df2['mean'], bar_width, label='Average Power', color='b', align='center')
plt.bar([i + bar_width for i in index], agg_df2['max'], bar_width, label='Max Power', color='g', align='center')
plt.bar([i + 2 * bar_width for i in index], agg_df2['min'], bar_width, label='Min Power', color='r', align='center')

# Customize the graph
plt.xlabel('Car Body Type')
plt.ylabel('Power(in bhp)')
plt.title('Power Statistics by Car Body Type')
plt.xticks([i + bar_width for i in index], agg_df2['body_type'])
plt.legend()


#to make the third graph that is Comparision of Mileage
plt.subplot(3,3,7)
# Calculate the average, maximum, and minimum mileage for each car body type
agg_df3 = df.groupby("body_type")["mileage"].agg(['mean', 'max', 'min']).reset_index()

# Create the side-by-side bar graph
bar_width = 0.2
index = range(len(agg_df3))

plt.bar(index, agg_df3['mean'], bar_width, label='Average Mileage', color='b', align='center')
plt.bar([i + bar_width for i in index], agg_df3['max'], bar_width, label='Max Mileage', color='g', align='center')
plt.bar([i + 2 * bar_width for i in index], agg_df3['min'], bar_width, label='Min Mileage', color='r', align='center')

# Customize the graph
plt.xlabel('Car Body Type')
plt.ylabel('Mileage(in kmpl)')
plt.title('Mileage Statistics by Car Body Type')
plt.xticks([i + bar_width for i in index], agg_df3['body_type'])
plt.legend()

plt.tight_layout()
plt.show()   #to show all the graphs


'''Analysis
Here we can see that in all three parameters that is engine, Power and Mileage.
Average of all are nearly same so we will not consider this.
We will further look into these parameters using "fuel type" and "transmission type".'''
```

```python
#To compare car body type and fuel type with all three parameters that is Engine, Power and Mileage


#this is the first graph that is comparision of Mileage
# Calculate the average mileage for each combination of car body type and fuel type
average_mileage = df.groupby(['body_type', 'fuel'])['mileage'].mean().unstack()

# Plot the side-by-side bar graph
ax = average_mileage.plot(kind='bar', stacked=False, figsize=(10, 6))
ax.set_xlabel('Car Body Type')
ax.set_ylabel('Average Mileage(in kmpl)')
plt.xticks(rotation=0)
plt.title('Average Mileage by Car Body Type and Fuel Type')
plt.legend(title='Fuel Type', bbox_to_anchor=(1.05, 1), loc='upper left')

#to show the graph
plt.tight_layout()
plt.show()


'''Analysis
CNG cars have more mileage in all car body type, but count of them is less.
Diesel cars have more mileage than Petrol cars.'''
```

Hypothesis Testing

```python
#Below is the Z-test for hypothesis that number of cars launched in 2019 are the highest.


from scipy import stats
# Extract the launch years into a variable
launch_years = df['year']

# Count the occurrences of the year 2019
count_2019 = sum(year == 2019 for year in launch_years)

# Total number of samples
sample_size = 1000

# Proportion of cars launched in 2019
p = count_2019 / sample_size

# Expected proportion (if distribution is uniform across years)
expected_p = 1 / len(launch_years.unique())

# Standard deviation (assuming a binomial distribution)
std_dev = (expected_p * (1 - expected_p) / sample_size) ** 0.5

# Perform the Z-test
z_score = (p - expected_p) / std_dev

# Calculate the p-value (two-tailed test)
p_value = stats.norm.cdf(z_score) * 2

# Set significance level
alpha = 0.05

# Print the z-score and the p-value
print("Z-score:", z_score)
print("P-value:", p_value)

# Decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis: The number of cars launched in 2019 is significantly different.")
else:
    print("Fail to reject the null hypothesis: The number of cars launched in 2019 is not significantly different.")
```

```python
#Below is the chi-square test between the two groups, car body type and fuel type to check whether there is a significant difference between them or not.


from scipy.stats import chi2_contingency
# Extract the columns into separate variables
car_body_type = df['body_type']
fuel_type = df['fuel']

# Create a contingency table (cross-tabulation)
contingency_table = pd.crosstab(car_body_type, fuel_type)

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)

# Set significance level
alpha = 0.05

# Decision based on the p-value
if p < alpha:
    print("Reject the null hypothesis: There is a significant difference between car body type and fuel type.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between car body type and fuel type.")
```

```python
#Below is the one way ANOVA test of the three different types of cars, Jeep, SUV and Sedan to check whether there is a significant difference between them
#in performance or not.


from scipy.stats import f_oneway
# Extract the columns into separate variables
car_type = df['body_type']
mileage_group = df['max_power']

# Create subsets for each group/category from the 'mileage_group' column
group1 = df[df['body_type'] == 'Jeep']['max_power']
group2 = df[df['body_type'] == 'SUV']['max_power']
group3 = df[df['body_type'] == 'Sedan']['max_power']

# Perform the ANOVA test
f_stat, p_value = f_oneway(group1, group2, group3)

# Print the results
print("F-statistic:", f_stat)
print("P-value:", p_value)

# Set significance level
alpha = 0.05

# Decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between at least one pair of means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")
```
