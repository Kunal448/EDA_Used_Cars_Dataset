```python
df['engine'] = df['engine'].str.replace(r'\D', '', regex=True).astype('Int64')
```

```python
df['engine']=np.where(df['engine'].isna(),df['engine'].mean(),df['engine'])
```

```python
df["number"] = [x for x in range(1, 8129, 1)]
```

```python
first_column = df.pop('number')
df.insert(0, 'number', first_column)
```

```python
df['engine'] = df['engine'].astype('int64')
```

```python
df['seats'] = df['seats'].astype('Int64')
```

```python
df.drop('max_power',axis=1, inplace=True)
```

```python
import random
def calculate_power(engine):
    if 500 <= engine <= 600:
        return random.randint(39, 51)
    elif 601 <= engine <= 700:
        return random.randint(46, 54)
    elif 701 <= engine <= 800:
        return random.randint(53, 61)
    elif 801 <= engine <= 900:
        return random.randint(59, 111)
    elif 901 <= engine <= 1000:
        return random.randint(79, 141)
    elif 1001 <= engine <= 1200:
        return random.randint(99, 151)
    elif 1201 <= engine <= 1500:
        return random.randint(129, 201)
    elif 1501 <= engine <= 1800:
        return random.randint(159, 241)
    elif 1801 <= engine <= 2000:
        return random.randint(199, 301)
    elif 2001 <= engine <= 2500:
        return random.randint(269, 341)
    elif 2501 <= engine <= 3000:
        return random.randint(309, 401)
    elif 3001 <= engine <= 3500:
        return random.randint(379, 451)
    elif 3501 <= engine:
        return random.randint(399, 501)
    else:
        return None  # or any default value you prefer

# Apply the function to create the 'torque' column
df['max_power'] = df['engine'].apply(calculate_power)
```

```python
df.drop('torque',axis=1, inplace=True)
```

```python
#df['mileage'] = df['mileage'].str.replace(' kmpl', '').astype(float)
```

```python
df['mileage'] = df['mileage'].str.extract('([\d.]+)').astype(float)
```

```python
df['mileage']=np.where(df['mileage'].isna(),df['mileage'].mean(),df['mileage'])
```

```python
df['seats'] = df['seats'].astype('Int64')
```

```python
df.info()
```

```python
def generate_random_integer():
    return np.random.randint(4, 8 + 1)

# Fill null values with random integer values
df['seats'] = df['seats'].fillna(generate_random_integer())
```
